```python
import numpy as np
import time

def Hypothesis(W,b,X):
	H=np.dot(W.T,X)+b
	return H



def Cost_Function(X,Y,W,b,m):
    H= Hypothesis(W,b,X)
    J= 1/2/m*np.sum((Y-H)**2)
    return J

def Cost_Function_Derivative(X,Y,W,b,m):
	H= Hypothesis(W,b,X)
	derv=Y-H
	dw=-(1/m)*np.dot(derv,X.T)
	db=-(1/m)*np.sum(derv)
	return dw,db

def Gradient_Descent(X,Y,W,b,m,alpha):

	# j= Cost_Function(X,Y,W,b,m)

	dw,db=Cost_Function_Derivative(X,Y,W,b,m)

	W=W-alpha*(dw.T)
	b=b-alpha*db

	return W,b

def Linear_Regression(X,Y,W,b, m,alpha,num_iters):
	m = X.shape[1]
	for x in range(num_iters):
		W,b = Gradient_Descent(X,Y,W,b, m, alpha)
	
		if x % 100 == 0:
			print("\nIteration no:",str(x))
			print('W '+str(W))
			print("b "+str(b))
			print ('cost is ' +str(Cost_Function(X,Y,W,b,m)))
	return W,b


X1 = [[1,1],[1,2],[1,3],[1,4],[1,5],[1,6],[1,7],[1,8],[1,9]]
nx = len(X1[0])
m = len(X1)
X = np.zeros(shape=(nx, m))
for i in range(nx):
    for j in range(m):
        X[i][j] = X1[j][i]

#Vecor Y
Y = [19, 20, 20.5, 21.5, 22, 23, 23, 25.5, 24]
ylen = len(Y)
Y = np.array(Y)
Y = Y.reshape(1, ylen)


W=np.random.randn(nx, 1)

b = np.random.randn(1)[0]
print("Initial Values")
print('w=',str(W))
print('b',str(b))

alpha=0.01
start=time.time()
newW,newb=Linear_Regression(X,Y,W,b,len(Y),alpha,2000)
stop=time.time()

print("\nFinal values\n")
print('w=',str(newW))
print('b',str(newb))

print('Time taken:', round(1000*(stop-start), 2), 'ms')
```
